```python
import json

with open('/content/drive/MyDrive/APP1/train-v2.0.json', 'r') as f:
    data = json.load(f)

output_dict = {"data": []}

for article in data['data']:
    title = article['title']
    for paragraph in article['paragraphs']:
        context = paragraph['context']
        for qa in paragraph['qas']:
            question_text = qa['question']
            question_id = qa['id']
            for answer in qa['answers']:
                answer_text = answer['text']
                answer_start = answer['answer_start']
                output = {
                    "title": title,
                    "paragraphs": [
                        {
                            "context": context,
                            "qas": [
                                {
                                    "question": question_text,
                                    "id": question_id,
                                    "answers": [
                                        {
                                            "text": answer_text,
                                            "answer_start": answer_start
                                        }
                                    ]
                                }
                            ]
                        }
                    ]
                }
                
                output_dict["data"].append(output)



# write the dictionary to a file
with open('/content/drive/MyDrive/APP1/output.json', "w", encoding='utf-8') as outfile:
  json.dump(output_dict, outfile, ensure_ascii=False)
```
